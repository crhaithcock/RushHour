```python
import os
import pandas as pd
import networkx as nx
import matplotlib as plt
from matplotlib.collections import LineCollection
%matplotlib inline
```

```python
G = nx.Graph()
G.add_nodes_from([1,2,3,4,5,6])
G.add_edge(1,2)
G.add_edge(2,4)
G.add_edge(3,4)
G.add_edge(1,4)
G.add_edge(4,5)
G.add_edge(2,6)
G.add_edge(4,6)
G.add_edge(2,5)
pos = {1:[1,2],2:[2,1],3:[2,2],4:[2,3],5:[3,3],6:[3,2]}
```

```python
nx.draw(G,pos=pos,node_color=['r','b','b','b','b','b'],with_labels = ['1','2','3','4','5','6'])
```

```python
nx.draw_networkx(G,pos=pos,node_color=['r','b','b','b'])
```

```python
nx.draw_networkx_nodes(G, pos,nodecolor={})
edge_collection = nx.draw_networkx_edges(G, pos)
```
